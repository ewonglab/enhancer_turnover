```python
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc
from sklearn.metrics import roc_auc_score
import csv
import sys
import numpy as np
import pandas as pd
```

```python
def ro_curve(y_pred, y_label, figure_file, method_name):
    '''
        y_pred is a list of length n.  (0,1)
        y_label is a list of same length. 0/1
        https://scikit-learn.org/stable/auto_examples/model_selection/plot_roc.html#sphx-glr-auto-examples-model-selection-plot-roc-py  
    '''
    y_label = np.array(y_label)
    y_pred = np.array(y_pred)    
    fpr = dict()
    tpr = dict() 
    roc_auc = dict()
    fpr[0], tpr[0], _ = roc_curve(y_label, y_pred)
    roc_auc[0] = auc(fpr[0], tpr[0])
    lw = 2
    plt.plot(fpr[0], tpr[0],
         lw=lw, label= method_name + ' (area = %0.2f)' % roc_auc[0])
    plt.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    # plt.xticks(font="Times New Roman",size=18,weight="bold")
    # plt.yticks(font="Times New Roman",size=18,weight="bold")
    fontsize = 14
    plt.xlabel('False Positive Rate', fontsize = fontsize)
    plt.ylabel('True Positive Rate', fontsize = fontsize)
    #plt.title('Receiver Operating Characteristic Curve', fontsize = fontsize)
    plt.legend(loc="lower right")
    plt.savefig('/home/114/xz3123/hg_Chr2mmtrained.pdf')
    return 
```

```python
#draw RO-curver on test hg_Chr2 hgtrained both CEBPA & Hnf4a
y_pred_CEBPA = np.load("/g/data/zk16/xzhang/cross-species-domain-adaptation/model_out/CEBPA_HumanChr2_hgtrained.npy")
y_pred_Hnf4a = np.load("/g/data/zk16/xzhang/cross-species-domain-adaptation/model_out/Hnf4a_HumanChr2_hgtrained.npy")

file_path_C = "/g/data/zk16/xzhang/cross-species-domain-adaptation/model_out/test_chr2_label/CEBPA_hg_lable.csv"
label_C = np.loadtxt(file_path_C)
y_label_C = label_C[0:4176896]

file_path_H = "/g/data/zk16/xzhang/cross-species-domain-adaptation/model_out/test_chr2_label/Hnf4a_hg_lable.csv"
label_H = np.loadtxt(file_path_H)
y_label_H = label_H[0:4176896]

ro_curve(y_pred_CEBPA ,y_label_C,"name","CEBPA")
ro_curve(y_pred_Hnf4a ,y_label_H,"name","Hnf4a")

```
