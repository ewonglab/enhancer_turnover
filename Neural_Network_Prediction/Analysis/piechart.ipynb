```python
#the Piechat-plot Mouse_enhancers by hg_trained and mm_trained DA models
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
from itertools import chain
from scipy.stats import *

from pylab import *
subplots_adjust(left=0.0,bottom=0.0,top=1.5,right=1.5)
# plt.style.use('ggplot')
fig, axes = plt.subplots(2,2)
axes = axes.flatten()
i =0
title = ["M_enhancers_Func_by_mmTrained","hg38_nonFunc_by_mmTrained","M_enhancers_Func_by_hgTrained","hg38_nonFunc_by_hgTrained"]


# data preparation
# CBEPA
a = np.load("/g/data/zk16/xzhang/cross-species-domain-adaptation/model_out/CEBPA_DA-mmtrained_mm10-test_ME.preds.npy")
b = np.load("/g/data/zk16/xzhang/cross-species-domain-adaptation/model_out/CEBPA_DA-mmtrained_hg38-test_ME.preds.npy")
ta = np.load("/g/data/zk16/xzhang/cross-species-domain-adaptation/model_out/Hnf4a_DA-mmtrained_mm10-test_ME.preds.npy")
tb = np.load("/g/data/zk16/xzhang/cross-species-domain-adaptation/model_out/Hnf4a_DA-mmtrained_hg38-test_ME.preds.npy")
# Hnf4a
c = np.load("/g/data/zk16/xzhang/cross-species-domain-adaptation/model_out/CEBPA_DA-hgtrained_mm10-test_ME.preds.npy")
d = np.load("/g/data/zk16/xzhang/cross-species-domain-adaptation/model_out/CEBPA_DA-hgtrained_hg38-test_ME.preds.npy")
tc = np.load("/g/data/zk16/xzhang/cross-species-domain-adaptation/model_out/Hnf4a_DA-hgtrained_mm10-test_ME.preds.npy")
td = np.load("/g/data/zk16/xzhang/cross-species-domain-adaptation/model_out/Hnf4a_DA-hgtrained_hg38-test_ME.preds.npy")


la = list(chain.from_iterable(a))
lb = list(chain.from_iterable(b))
tla = list(chain.from_iterable(ta))
tlb = list(chain.from_iterable(tb))

lc = list(chain.from_iterable(c))
ld = list(chain.from_iterable(d))
tlc = list(chain.from_iterable(tc))
tld = list(chain.from_iterable(td))

#mouse Func _by mmtrained model
index_a = [index for (index, number) in enumerate(la) if number >= 0.9]
index_ta = [index for (index, number) in enumerate(tla) if number >= 0.9]
result = list(set(index_a).union(set(index_ta)))
value = [len(result),(len(a)-len(result))]
print(value)
#plot
colors = ['pink','lightblue']
labels = ['bound','unbound']
explode = (0.1, 0)  
plt.sca(axes[0]) 
plt.pie(x = value, 
        labels = labels,
        colors = colors,
        explode = explode,
        autopct = '%.1f%%')
plt.title(title[0],fontsize=10)

#mouse NONFunc _by mmtrained model
index_a = [index for (index, number) in enumerate(lb) if number >= 0.9]
index_ta = [index for (index, number) in enumerate(tlb) if number >= 0.9]
result = list(set(index_a).union(set(index_ta)))
value = [len(result),(len(a)-len(result))]
print(value)
#plot
colors = ['pink','lightblue']
labels = ['bound','unbound']
explode = (0.1, 0)  
plt.sca(axes[1]) 
plt.pie(x = value, 
        labels = labels,
        colors = colors,
        explode = explode,
        autopct = '%.1f%%')
plt.title(title[1],fontsize=10)

#mouse Func _by hgtrained model
index_a = [index for (index, number) in enumerate(lc) if number >= 0.9]
index_ta = [index for (index, number) in enumerate(tlc) if number >= 0.9]
result = list(set(index_a).union(set(index_ta)))
value = [len(result),(len(a)-len(result))]
print(value)
#plot
colors = ['pink','lightblue']
labels = ['bound','unbound']
explode = (0.1, 0)  
plt.sca(axes[2]) 
plt.pie(x = value, 
        labels = labels,
        colors = colors,
        explode = explode,
        autopct = '%.1f%%')
plt.title(title[2],fontsize=10)

#mouse NONFunc _by hgtrained model
index_a = [index for (index, number) in enumerate(ld) if number >= 0.9]
index_ta = [index for (index, number) in enumerate(tld) if number >= 0.9]
result = list(set(index_a).union(set(index_ta)))
value = [len(result),(len(a)-len(result))]
print(value)
#plot
colors = ['pink','lightblue']
labels = ['bound','unbound']
explode = (0.1, 0)  
plt.sca(axes[3]) 
plt.pie(x = value, 
        labels = labels,
        colors = colors,
        explode = explode,
        autopct = '%.1f%%')
plt.title(title[3],fontsize=10)

plt.savefig('/home/114/xz3123/MouseEnhancers_0.9.pdf')   
plt.show()
```
